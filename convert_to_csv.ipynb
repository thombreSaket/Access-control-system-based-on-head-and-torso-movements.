```python
import os
import csv

# define function to convert txt files to csv
def txt_to_csv(txt_file_path):
    csv_file_path = os.path.splitext(txt_file_path)[0] + '.csv'
    with open(txt_file_path, 'r') as txt_file, open(csv_file_path, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        for line in txt_file:
            csv_writer.writerow(line.strip().split())
    return csv_file_path

# define function to process a folder
def process_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            txt_file_path = os.path.join(folder_path, filename)
            csv_file_path = txt_to_csv(txt_file_path)
            os.remove(txt_file_path)
            print(f'{filename} converted to {csv_file_path} and deleted.')

# specify root folder
root_folder = r'C:\Users\sudha\Desktop\Semester 2\MLS\Project 2\OneDrive_2023-04-15\Data of Head and Torso movement\data\\'


# loop through subfolders and process each one
for subdir, dirs, files in os.walk(root_folder):
    for folder_name in dirs:
        folder_path = os.path.join(subdir, folder_name)
        process_folder(folder_path)
```
